#  Missing value

```python
# Inspecting the missing values in the dataset
print(df.isnull().values.sum())

# Replacing the ' 's with NaN
df = df.replace(" ",np.NaN)

# Impute the missing values with mean imputation
df = df.fillna(df.mean(numeric_only=True))

# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

```python
print(df.isnull())
```

# Theorem
Solve missing
- Ostrict theorem - remove the whole row if a data is missing from it. (Not recommended)
- Mean imputation - if any values are missing we take mean of all the values and fill the missing values with the mean values. (Used on field with loose requirements for accurate data. ie. not used in sensitive data example for medical practices)

```python
# Replace the ''(null) values with NaN
df = df.replace(" ",np.NaN)          # if there is missing values

# Impute the missing values with mean imputation
df = df.fillna(df.mean())                         # Mean Theorem

# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

```python
# Removing the column having entire missing value
df = df.dropna(axis = 1, how ='all')
```

```python
# Replace the ' 's with NaN
df = df.replace(" ",np.NaN)     # If there is missing values
```

```python
# Since the missing values in not numerical value we are going to replace the value with a constant 0.
df = df.fillna(0)
```

```python
# Reinspecting the missing value in dataset
print(df.isnull().values.sum())
```

When the data is not missing any values we are good to go for EDA cycle

# EDA CYCLE

### Histogram

```python
# Checking the data type as Histogram excepts either integer and float

print(df['age'].dtype)
print(df['smoke'].dtype)
```

```python
sns.histplot(df['age'], bins = 12, color = 'lightgreen', kde = False, edgecolor = 'green')


# Add labels
plt.title('Smoke with age')
plt.xlabel('age')
plt.ylabel('smoke')

# Show the plot
plt.show()
```

- It is right skewed.

```python
# With kde
sns.histplot(df['age'], bins = 10, color = 'lightgreen', kde = True, edgecolor = 'green')


# Add labels
plt.title('Smoke with age')
plt.xlabel('age')
plt.ylabel('smoke')

# Show the plot
plt.show()
```

### Scatterplot

```python
# Checking the data type as scatterplot excepts either integer and float
print(df['age'].dtype)
print(df['smoke'].dtype)
```

```python
# Seaborn Scatterplot     - mostly used if we have numerical values on both x and y and want to see the relation between them
sns.scatterplot(x=df['smoke'], y=df['age'])  # hue shows U- urban and R-rural


# Optionally, add labels and title
plt.title('Scatterplot of age and smoke')    # Overwrites already shown output in matplotlib. 
plt.xlabel('smoke')                                 # Shows without these as well. And they both work on same engine.
plt.ylabel('age')                             # Engine - seaborn --- accessed by sns.

#Show the plot           matplotlib covered one.
plt.show()
```

```python
# Scatterplot with hue 
sns.scatterplot(x=df['smoke'], y=df['age'], hue =df['age'])  # hue displays the weight

# Move the legend outside the plot area so it does no obstruct the view.
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.title('Scatterplot of age and smoke')    
plt.xlabel('smoke')
plt.ylabel('age')                        

plt.show()
```

```python
# seaborn histogram 
sns.distplot(df['age'], hist=True, kde=False, 
             bins=12, color = 'blue',
             hist_kws={'edgecolor':'black'})
# Adding labels
plt.title('age count')
plt.xlabel('age ')

plt.ylabel('Count')
```

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', size = 15)
colormap = sns.diverging_palette(10, 220, as_cmap = True)
sns.heatmap(df.corr(),
            cmap = colormap,
            square = True,
            annot = True,
            linewidths=0.1,vmax=1.0, linecolor='white',
            annot_kws={'fontsize':12 })
plt.show()
```
